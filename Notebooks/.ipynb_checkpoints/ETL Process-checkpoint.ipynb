## Lectura de archivos

```python
import pandas as pd
dfEvents = pd.read_csv('../Data/events.csv')
dfEvents
```

```python
dfCategoria = pd.read_csv('../Data/categoria.csv')
dfCategoria
```

```python
dfProducto = pd.read_csv('../Data/producto.csv')
dfProducto
```

## Creacion de dimensiones

### Dimension Marca

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+mysqldb://yadderace:150592@localhost:3306/DBAppliedDS')

dimMarca = dfMarca[['id', 'marca']]
dimMarca.rename(columns ={'id': 'id_marca', 'marca': 'nombre_marca'}, inplace = True)
dimMarca['nombre_marca'] = dimMarca['nombre_marca'].str.upper()

dimMarca.to_sql('dim_marca', index = False, index_label='id_marca', if_exists = 'replace', con = engine)

dimMarca
```

### Dimension Categoria

```python
dimCategoria = dfCategoria[['id', 'categoria']]
dimCategoria.rename(columns ={'id': 'id_categoria', 'categoria': 'nombre_categoria'}, inplace = True)
dimCategoria['nombre_categoria'] = dimCategoria['nombre_categoria'].str.upper().str.strip()

dimCategoria.to_sql('dim_categoria', index = False, index_label='id_categoria', if_exists = 'replace', con = engine)

dimCategoria
```

### Dimension Producto

```python
dimProducto = pd.DataFrame(dfProducto[['id', 'nombre']])
dimProducto.rename(columns ={'id': 'id_producto', 'nombre': 'nombre_producto'}, inplace = True)
dimProducto.drop_duplicates(subset = ['id_producto'], keep = 'last', inplace = True)
dimProducto['nombre_producto'] = dimProducto['nombre_producto'].str.upper().str.strip()

dimProducto['nombre_producto'] = [nombre.replace("\u032c", "") for nombre in dimProducto['nombre_producto']]

dimProducto.to_sql('dim_producto', index = False, index_label='id_producto', if_exists = 'replace', con = engine)

dimProducto
```

### Dimension Idioma

```python
dimIdioma = pd.DataFrame(dfCliente[['idioma']])
dimIdioma.drop_duplicates(subset = ['idioma'], inplace = True)
dimIdioma.reset_index(inplace = True, drop = True)
dimIdioma['id_idioma'] = dimIdioma.index
dimIdioma.rename(columns ={'idioma': 'nombre_idioma'}, inplace = True)
dimIdioma['nombre_idioma'] = dimIdioma['nombre_idioma'].str.upper().str.strip()
dimIdioma['nombre_idioma'] = [str(bytes(nombre, 'utf-8')\
                                  .decode('utf-8','ignore')\
                                  .encode("utf-8"))\
                                  .replace('b','').replace('\'','')
                              for nombre in dimIdioma['nombre_idioma']]

dimIdioma.to_sql('dim_idioma', index = False, index_label='id_idioma', if_exists = 'replace', con = engine)

dimIdioma
```

### Dimension Ciudad

```python
dimCiudad = pd.DataFrame(dfCliente[['ciudad']])
dimCiudad.drop_duplicates(subset = ['ciudad'], inplace = True)
dimCiudad.reset_index(inplace = True, drop = True)
dimCiudad['id_ciudad'] = dimCiudad.index
dimCiudad.rename(columns ={'ciudad': 'nombre_ciudad'}, inplace = True)
dimCiudad['nombre_ciudad'] = dimCiudad['nombre_ciudad'].str.upper().str.strip()
dimCiudad['nombre_ciudad'] = [str(bytes(nombre, 'utf-8')\
                                  .decode('utf-8','ignore')\
                                  .encode("utf-8"))\
                                  .replace('b','').replace('\'','')
                              for nombre in dimCiudad['nombre_ciudad']]

dimCiudad.to_sql('dim_ciudad', index = False, index_label='id_ciudad', if_exists = 'replace', con = engine)

dimCiudad
```

### Dimension  Evento

```python
dimEvento = pd.DataFrame(dfEvents[['event']])
dimEvento.drop_duplicates(subset = ['event'], inplace = True)
dimEvento.reset_index(inplace = True, drop = True)
dimEvento['id_evento'] = dimEvento.index
dimEvento.rename(columns ={'event': 'nombre_evento'}, inplace = True)
dimEvento['nombre_evento'] = dimEvento['nombre_evento'].str.upper().str.strip()

dimEvento.to_sql('dim_evento', index = False, index_label='id_evento', if_exists = 'replace', con = engine)

dimEvento
```

### Dimension Tiempo

```python
import sqlalchemy

dimTiempo = pd.DataFrame(dfEvents[['timestamp']])
dimTiempo.drop_duplicates(subset = ['timestamp'], inplace = True)
dimTiempo['fecha_hora'] = pd.to_datetime(dimTiempo["timestamp"], unit='ms')
dimTiempo['fecha'] = dimTiempo['fecha_hora'].dt.date
dimTiempo['horas'] = dimTiempo['fecha_hora'].dt.time
dimTiempo['anio'] = pd.DatetimeIndex(dimTiempo['fecha']).year
dimTiempo['mes'] = pd.DatetimeIndex(dimTiempo['fecha']).month
dimTiempo['dia'] = pd.DatetimeIndex(dimTiempo['fecha']).day
dimTiempo['hora'] = dimTiempo['fecha_hora'].dt.hour
dimTiempo['minuto'] = dimTiempo['fecha_hora'].dt.minute
dimTiempo['segundo'] = dimTiempo['fecha_hora'].dt.second

dimTiempo.to_sql('dim_tiempo', index = False, index_label='timestamp', if_exists = 'replace', 
                  dtype = {'timestamp': sqlalchemy.types.BigInteger()}, con = engine)

dimTiempo
```

### Dimension Cliente

```python
dimCliente = pd.DataFrame(dfCliente[['id', 'nombre', 'apellido', 'nacimiento', 'genero', 'empresa', 'nit', 'puesto', 'correo', 'telefono']])
dimCliente['nacimiento'] = pd.to_datetime(dimCliente["nacimiento"])
dimCliente.rename(columns ={'id': 'id_cliente'}, inplace = True)

dimCliente.to_sql('dim_cliente', index = False, index_label='id_cliente', if_exists = 'replace', con = engine)

dimCliente
```

## Creacion de tabla de hechos

### Informacion Evento

```python
dfAccionView = dfEvents[['timestamp', 'visitorid', 'event', 'itemid', 'transactionid']]

# Agregando id_evento
dfAccionView.rename(columns = {'event': 'nombre_evento'}, inplace = True)
dfAccionView['nombre_evento'] = dfAccionView['nombre_evento'].str.upper()
dfAccionView = dfAccionView.join(dimEvento.set_index('nombre_evento'), on = 'nombre_evento', how = 'left')
dfAccionView.drop(columns = ['nombre_evento'], inplace = True)

dfAccionView
```

### Informacion Cliente

```python
# Agregando id_cliente y la info de cliente para agregar mas dimensiones (idioma, ciudad)
dfAccionView.rename(columns = {'visitorid': 'id_cliente'}, inplace = True)
# Agregando idioma
dfAccionView = dfAccionView.join(dfCliente[['id','idioma', 'ciudad']].set_index('id'), on = 'id_cliente', how = 'left')
dfAccionView['idioma'] = dfAccionView['idioma'].str.upper()
dfAccionView = dfAccionView.join(dimIdioma.set_index('nombre_idioma'), on = 'idioma', how = 'left')
# Agregando ciudad
dfAccionView['ciudad'] = dfAccionView['ciudad'].str.upper()
dfAccionView = dfAccionView.join(dimCiudad.set_index('nombre_ciudad'), on = 'ciudad', how = 'left')

dfAccionView.drop(columns = ['idioma', 'ciudad'], inplace = True)

dfAccionView
```

### Informacion Producto

```python
dfProducto['id'].value_counts(dropna = False)

dfProducto.loc[dfProducto['id'] == 356475]
```

```python
dfProducto.drop_duplicates(subset = ['id'], keep = 'last', inplace = True)
dfAccionView = dfAccionView.join(dfProducto.set_index('id'), on = 'itemid', how = 'left')
dfAccionView.rename(columns = {'itemid': 'id_producto', 
                               'transactionid': 'id_transaction', 
                               'categoria_id':'id_categoria',
                               'marca_id': 'id_marca'
                              }, inplace = True)
dfAccionView.drop(columns = ['nombre'], inplace = True)
dfAccionView[['volumen', 'precio']] = dfAccionView[['volumen', 'precio']].fillna(value = 0)
dfAccionView
```

### Registro BD

```python
import sqlalchemy

dfAccionView.to_sql('fact_eventos', index = True, if_exists = 'replace', dtype = {
    'timestamp': sqlalchemy.types.BigInteger(),
    'id_cliente':sqlalchemy.types.BigInteger(),
    'id_producto':sqlalchemy.types.BigInteger(),
    'id_transaction': sqlalchemy.types.BigInteger(),
    'id_evento': sqlalchemy.types.BigInteger(),
    'id_idioma': sqlalchemy.types.BigInteger(),
    'id_ciudad': sqlalchemy.types.BigInteger(),
    'id_categoria': sqlalchemy.types.BigInteger(),
    'id_marca': sqlalchemy.types.BigInteger()
}, con = engine)
```
